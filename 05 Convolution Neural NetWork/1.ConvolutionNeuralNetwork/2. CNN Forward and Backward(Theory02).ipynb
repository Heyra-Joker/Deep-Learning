###  Forward Propagation

$Z^{[1]}=CONV1(W^{[1]},X)+b^{[1]}$
- $X=A^{[0]} shape:(32\times 32\times 3),W^{[1]} shape:(5\times 5\times 3\times 6),b^{[1]} shape:(1\times 1\times 6),Z^{[1]}shape:(28\times 28 \times 6)$

$A^{[1]}=ReLu(Z^{[1]})$
- $A^{[1]}shape:(28\times 28 \times 6)$

$A^{[1]}=Pool1(P^{[1]},A^{[1]})$
- $P^{[1]}shape:(2\times 2\times 6),A^{[1]}shape:(28\times 28 \times 6),A^{[1]}shape(14\times 14\times 6)$

$Z^{[2]}=CONV2(W^{[2]},A^{[1]})+b^{[2]}$
- $A^{[1]} shape:(14\times 14\times 6),W^{[2]} shape:(5\times 5\times 6\times 16),b^{[2]} shape:(1\times 1\times 16),Z^{[2]}shape:(10\times 10 \times 16)$

$A^{[2]}=ReLu(Z^{[2]})$
- $Z^{[2]}shape:(10\times 10 \times 16),A^{[2]}shape:(10\times 10 \times 16)$

$A^{[2]}=Pool2(P^{[2]},A^{[2]})$
- $P^{[2]}shape:(2\times 2\times 16),A^{[2]}shape(10\times 10\times 16),A^{[2]}shape(5\times 5\times 16)$

Flatten $A^{[2]}$
- $A^{[2]} shape:(400\times 1)$

$Z^{[3]} = W^{[3]}*A^{[2]}+b^{[3]}$
- $A^{[2]} shape:(400\times 1),W^{[3]} shape:(120\times 400),b^{[3]} shape:(120 \times 1),Z^{[3]}shape:(120\times 1)$

$A^{[3]}=ReLu(Z^{[3]})$
- $A^{[3]}shape:(120\times 1)$

$Z^{[4]} = W^{[4]}*A^{[3]}+b^{[4]}$
- $A^{[3]} shape:(120\times 1),W^{[4]} shape:(84\times 120),b^{[4]} shape:(84 \times 1),Z^{[4]}shape:(84\times 1)$

$A^{[4]}=ReLu(Z^{[4]})$

$Z^{[5]} = W^{[5]}*A^{[4]}+b^{[5]}$

$J=Loss(A^{[5]},labels)$

### Backward Propagation

这里我们直接从$dZ^{[5]}$开始:



$dZ^{[5]}=A^{[5]}-y^{[true]}$

- $dZ^{[5]}shape:(10\times 1),A^{[5]}shape:(10\times 1),y^{[true]}shape:(10\times 1)$

$dW^{[5]}=dZ^{[5]}A^{[4]^{T}}$

- $dW^{[5]}shape:(10\times 84),dZ^{[5]}shape:(10\times 1),A^{[4]^{T}}shape:(1\times 84)$

$db^{[5]}=dZ^{[5]}$

- $db^{[5]}shape:(10\times 1),dZ^{[5]}shape:(10\times 1)$

$dA^{[4]}=W^{[5]^{T}}dZ^{[5]}$

- $dA^{[4]}shape:(84\times 1),W^{[5]^{T}}shape:(84\times 10),Z^{[5]}shape:(10\times 1)$

$dZ^{[4]}=dA^{[4]}\odot g^{'}(Z^{[4]})$

- $dZ^{[4]}shape:(84\times1),dA^{[4]}shape:(84\times1),Z^{[4]}shape:(84\times1)$

$dW^{[4]}=dZ^{[4]}A^{[3]^{T}}$

- $dW^{[4]}shape:(84\times 120),dZ^{[4]}shape:(84\times1),A^{[3]^{T}}shape:(1\times 120)$

$db^{[4]}=dZ^{[4]}$

- $db^{[4]}shape:(84\times 1),dZ^{[4]}shape:(84\times1)$

$dA^{[3]}=W^{[4]^{T}}dZ^{[4]}$

- $dA^{[3]}shape:(120\times1),W^{[4]^{T}}shape:(120\times 84),dZ^{[4]}shape:(84\times1)$

$dZ^{[3]}=dA^{[3]}\odot g^{'}(Z^{[3]})$

- $dZ^{[3]}shape:(120\times1),dA^{[3]}shape:(120\times 1),Z^{[3]}shape:(120\times1)$

$dW^{[3]}=dZ^{[3]}A^{[2]^{T}}$

- $dW^{[3]}shape:(120\times 400),dZ^{[3]}shape:(120\times1),A^{[2]^{T}}shape:(1\times 400)$

$db^{[3]}=dZ^{[3]}$

- $db^{[3]}shape:(120\times 1),dZ^{[3]}shape:(120\times1)$

$dA^{[2]}=W^{[3]^{T}}dZ^{[3]}$

- $dA^{[2]}shape:(400\times1),W^{[3]^{T}}shape:(400\times 120),dZ^{[4]}shape:(120\times1)$

$dA^{[2]} \;reshape$

- $dA^{[2]}shape:(5\times 5\times 16)$

$dMaxPoold:dA^{[2]}=dA^{[2]}$

- $dA^{[2]}shape:(10\times10\times16),dA^{[2]}shape:(5\times5\times16)$

$dZ^{[2]}=dA^{[2]}\odot g^{'}(Z^{[2]})$

- $dZ^{[2]}shape:(10\times10\times16),dA^{[2]}shape:(10\times10\times16),Z^{[2]}shape:(10\times10\times 16)$

$dW^{[2]}=A^{[1]}"*"dZ^{[2]}$

- $dW^{[2]}shape:(5\times5\times6\times16),A^{[1]}shape:(14\times14\times6),dZ^{[2]}shape:(10\times10\times16)$

$db^{[2]}=\sum_{row,col} dZ^{[2]}$

- $db^{[2]}shape:(1\times 1\times 16),dZ^{[2]}shape:(10\times 10\times 16)$

$dA^{[1]}=dZ^{[2]}"*"rot180(W^{[2]})$

- $dA^{[1]}shape:(14\times 14\times 6),dZ^{[2]}shape:(10\times 10\times 16),W^{[2]}shape:(5\times5\times 6\times 16)$
- $dZ^{[2]}brocasting(10\times 10\times 6) padding:4,Then\;dZ^{[2]} shape:(18\times 18\times 6)$,CONV $(5\times 5\times 6)$

$dMaxPoold:dA^{[1]}=dA^{[1]}$

- $dA^{[1]}shape:(28 \times 28\times 6),dA^{[1]}shape:(14\times 14\times 6)$

$dZ^{[1]}=dA^{[1]}\odot g^{'}(Z^{[1]})$

- $dZ^{[1]}shape:(28\times28\times6),dA^{[1]}shape:(28\times28\times6),Z^{[1]}shape:(28\times28\times 6)$

- $dW^{[1]}shape:(5\times5\times3\times 6),A^{[0]}=X\;shape:(32\times32\times3),dZ^{[1]}shape:(28\times28\times6)$
